```python
import os
import pandas as pd
from glob import glob

input_f = r'D:\가은\2021_DIYA\Attention\crawling'              # csv 파일이 있는 디렉토리 위치
output_f = r'D:\가은\2021_DIYA\Attention\crawling\region_result.csv'  # 병합하고 저장하려는 파일명

f_lst = glob(os.path.join(input_f, '*.csv'))                   # glob 함수로 csv로 시작하는 파일 불러오기

all_data = []                        # 불러온 csv 파일 내용을 저장할 빈 리스트
for f in f_lst:
    df = pd.read_csv(f)              # for문으로 csv 파일 읽기
    df.drop(columns=["Unnamed: 0"], inplace=True)
    df.dropna(axis=0, inplace=True)  # 전체: 111,144개, NaN: 19,971개
    all_data.append(df)              # 빈 리스트에 csv 내용 추가
    
dataCombine = pd.concat(all_data, axis=0, ignore_index=True) # axis=0: 수직 axis=1: 수평, ignore_index=True: 기존 순서를 무시하고 인덱스 재정렬
dataCombine.head(20)  # 91,173개
```

```python
# 중복 제거
df = dataCombine.copy()
df[df.duplicated(subset=['Rating','Category', 'Review'])] #  19019
df.drop_duplicates(subset=['Rating','Category', 'Review'], inplace=True) # 72154
df.reset_index(inplace=True)  # index 재설정
```
