```python
import pandas as pd
```

```python
# load data
name_list = ['0-479', '480-579', '580-598','busan_0-19','busan_20-101','jibang','jibang2']
paths = [f'diningcode_{name}.parquet' for name in name_list]

df_org = pd.DataFrame(columns=['Region','Rating','Category','Review'])
for path in paths:
    df_org = df_org.append(pd.read_parquet(path))
df_org.reset_index(inplace=True, drop=True)
```

```python
df_org
```

```python
# remove blank reviews
df_org.drop(index=df_org.loc[df_org['Review'] == ''].index, inplace=True)
df_org
```

```python
# remove duplicated rows
df = df_org.drop_duplicates(subset=['Rating','Review'], ignore_index=True)
df
```

```python
df['Rating'].value_counts()
```

```python
# 3-class
negative = [0,1,2]
neutral = [3]
positive = [4,5]

df['Rating'].replace(negative, 1, inplace=True)
df['Rating'].replace(neutral, 2, inplace=True)
df['Rating'].replace(positive, 3, inplace=True)
```

```python
df['Rating'].value_counts()
```

```python
df.to_parquet('diningcode.parquet')
```
