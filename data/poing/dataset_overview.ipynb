```python
import pandas as pd
from glob import glob
```

```python
paths = sorted(glob('./*.parquet'))
print(paths)
```

```python
paths = sorted(glob('./*.parquet'))
df = pd.DataFrame(columns=['Region','Rating','Category','Review'])
for path in paths:
    df = df.append(pd.read_parquet(path))
df.reset_index(inplace=True, drop=True)
```

```python
df
```

```python
# remove duplicated rows
df_unique = df.drop_duplicates(subset=['Rating','Review'], ignore_index=True)
df_unique
```

```python
df_unique['Rating'].value_counts()
```

```python
df_unique["Rating"] = (pd.to_numeric(df['Rating']))
df_unique.round({"Rating" : 0})
df_unique.astype({"Rating": int})
df_unique['Rating'].value_counts()
```

```python
df_unique[df_unique["Rating"] == 2.0]
```

```python
# 3-class
negative = [0.0, 1.0]
neutral  = [2.0, 3.0]
positive = [4.0, 5.0]

df_unique['Rating'].replace(negative, 1, inplace=True)
df_unique['Rating'].replace(neutral, 2, inplace=True)
df_unique['Rating'].replace(positive, 3, inplace=True)
```

```python
df_unique['Rating'].value_counts()
```

```python
df_unique.to_parquet("{}.parquet".format("POING"))
```
